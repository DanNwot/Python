# Mais sobre o return

### Pontos importantes:

- Podemos usar no return praticamente qualquer tipo de objeto: (número, string, lista, tupla, dicionário, outros objetos, etc.)
- O return, se for executado, encerra a função, mesmo que dentro dela haja um loop.

```python
#retornar um número
def minha_soma(num1, num2, num3):
    return num1 + num2 + num3

#retornar um texto
def padronizar_texto(texto):
    texto = texto.casefold()
    texto = texto.replace("  ", " ")
    texto = texto.strip()
    return texto

#retornar um boolean
def bateu_meta(vendas, meta):
    if vendas >= meta:
        return True
    else:
        return False
    
#retornar uma lista, tupla ou dicionario
def filtrar_lista_texto(lista, pedaco_texto):
    lista_filtrada = []
    for item in lista:
        if pedaco_texto in item:
            lista_filtrada.append(item)
            return lista_filtrada
```

```python
listaemails= ['daniel@gmail.com', 'cleber@hotmail.com', 'cleiton @gmail.com', 'jose@gmail.com']
def filtrar_emails_gmail(lista):
    lista_gmail = []
    for email in lista:
        if '@gmail.com' in email:
            lista_gmail.append(email)
    return lista_gmail
print(f'Os emails filtrados por gmail foram: {filtrar_emails_gmail(listaemails)}')

def filtrar_emails_hotmail(lista):
    lista_hotmail = []
    for email in lista:
        if '@hotmail.com' in email:
            lista_hotmail.append(email)
    return lista_hotmail
print(f'Os emails filtrados por hotmail foram: {filtrar_emails_hotmail(listaemails)}')
```
