## Oferecendo desconto para cliente com base na última compra

Suponha que você está trabalhando para uma empresa que deseja rastrear a atividade do cliente. Uma métrica que eles estão interessados é o tempo que passou desde a última transação do cliente. Se for muito tempo, eles podem oferecer um desconto para o cliente. Crie um script Python que mostra quanto tempo se passou desde a última compra do cliente. Se faz mais de 30 dias, mostre uma mensagem oferecendo um desconto para o cliente.

```python
from datetime import datetime
ultima_compra = datetime(2023, 5 ,10)
dataatual = datetime.now()
dias_desde_ultima_compra = (dataatual - ultima_compra).days
print(f"Já se passaram {dias_desde_ultima_compra} dias desde a última compra.")
if dias_desde_ultima_compra > 30:
    print("É hora de fazer uma nova compra!")
else:
    print("Ainda não é necessário fazer uma nova compra.")
```

## Data e hora em diferentes fusos horários

Uma empresa tem escritórios em São Paulo, Nova York e Tóquio. Crie um script Python que mostra a data e hora atuais nesses três fusos horários. Exiba, também, se estes escritórios estão abertos ou fechados (9h às 17h).


```python
from datetime import datetime
from zoneinfo import ZoneInfo
horasfuncionais = range(9, 18)  # Horário comercial das 9h às 17h

dataatualBR = datetime.now(ZoneInfo("America/Sao_Paulo"))
print("Hora atual em São Paulo:", dataatualBR.strftime("%H:%M:%S"))

dataatualNY = datetime.now(ZoneInfo("America/New_York"))
print("Hora atual em Nova York:", dataatualNY.strftime("%H:%M:%S"))

dataatualTokyo = datetime.now(ZoneInfo("Asia/Tokyo"))
print("Hora atual em Tóquio:", dataatualTokyo.strftime("%H:%M:%S"))


datas = [dataatualBR, dataatualNY, dataatualTokyo]
for data in datas:
    if data.hour in horasfuncionais:
        print(f"Em {data.tzinfo}, é dentro do horário comercial.")
    else:
        print(f"Em {data.tzinfo}, é fora do horário comercial.")
```
