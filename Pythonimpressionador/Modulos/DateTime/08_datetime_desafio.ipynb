## Calculando a idade

Um usuário fornece sua data de nascimento no formato "dd/mm/aaaa". Crie um script Python que calcula a idade do usuário.


```python
from datetime import datetime

datanascimentouser = input("Digite sua data de nascimento (dd/mm/aaaa): ")
datanascimentouser = datetime.strptime(datanascimentouser, "%d/%m/%Y")
dataatual = datetime.now()

mesnascimento = datanascimentouser.month
mesatual = dataatual.month

dianascimento = datanascimentouser.day
diaatual = dataatual.day

anonascimento = datanascimentouser.year
anoatual = dataatual.year

if mesnascimento < mesatual or (mesnascimento == mesatual and dianascimento <= diaatual):
    idade = anoatual - anonascimento
else:
    idade = anoatual - anonascimento - 1
print(f"Você tem {idade} anos.")
```
