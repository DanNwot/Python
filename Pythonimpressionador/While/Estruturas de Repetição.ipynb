````python
popa= 80000
popb= 200000
taxa_a= 0.03
taxa_b= 0.015
anos = 0
while popa < popb:
    popa += popa * taxa_a
    popb += popb * taxa_b
    anos += 1
print(f'Ser√£o necess√°rios {anos} anos para que a popula√ß√£o A ultrapasse ou iguale a popula√ß√£o B.')
````

#### 7. Fa√ßa um programa que pe√ßa para o usu√°rio inserir o faturamento dos √∫ltimos 5 meses (individualmente) e informe o faturamento total (soma) e o faturamento m√©dio por m√™s (m√©dia).

````python
faturamento= []
for i in range (5):
    valor = float(input(f'Informe o faturamento do {i + 1}¬∫: '))
    faturamento.append(valor)
faturamentomenor = min(faturamento)
print(f'O menor faturamento mensal foi: {faturamentomenor}')
totalfaturamento = sum(faturamento)
mediafaturamento = totalfaturamento / len(faturamento)
print(f'A m√©dia de faturamento mensal foi: {mediafaturamento}')
print(f'a soma do faturamento foi de :{totalfaturamento}')
````

#### 8. Fa√ßa um programa que consiga categorizar a idade das equipes de uma empresa. Fa√ßa um programa que pe√ßa para n pessoas a sua idade, ao final o programa devera verificar se a m√©dia de idade da equipe varia entre 0 e 25 (jovem) ,26 e 60 (s√™nior) e maior que 60 (idosa); e ent√£o, dizer se a equipe √© jovem, s√™nior ou idosa, conforme a m√©dia calculada.

````python
pessoa = []
while True:
    nome = input("Digite o nome da pessoa (ou 'sair' para encerrar): ")
    if nome.lower() == 'sair':
        break
    idade = int(input("Digite a idade da pessoa: "))
    pessoa.append((nome, idade))
    if idade < 25:
        print(f'{nome} √© jovem com a idade de {idade}.')
    elif 25 <= idade <= 60:
        print(f'{nome} √© senior com a idade de {idade}.')
    else:
        print(f'{nome} √© idoso com a idade de {idade}.')
print(f"Lista de pessoas cadastradas: {pessoa}")
````

#### 9. Numa elei√ß√£o existem tr√™s candidatos. Fa√ßa um programa que pe√ßa o n√∫mero total de eleitores. Pe√ßa para cada eleitor votar e ao final mostrar o n√∫mero de votos de cada candidato.

````python
candidatos = {'A': 0, 'B': 0, 'C': 0}

while True:
    votos = input('Digite em qual candidato quer votar (A, B, C) ou Sair para encerrar: ')
    if votos.lower() == 'sair':
        break
    elif votos.upper() not in candidatos:
        print('Candidato inv√°lido. Tente novamente.')
    else:
        candidatos[votos.upper()] += 1

print(f'Total de votos por candidato: {candidatos}')
````
